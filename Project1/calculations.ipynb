## Heuristic Nearest Neighbor

```python
import pandas as pd
import numpy as np
import tsplib95 as tsp
import matplotlib.pyplot as plt

file_names = ['A', 'B']
# Keeping data splitted into A and B problems
data = {}
# Structure for keeping coords of nodes, needed for plotting
nodes = {}


for fn in file_names:
    # Loading csv's
    data[fn] = pd.read_csv(f'data/results_kro{fn}100.csv', sep=',', header=None)
    data[fn] = data[fn].T
    
    # Loading problem
    problem = tsp.load(f'kro{fn}100.tsp')
    nodes_ = [v for _, v in problem.node_coords.items()]

    # Saving coords
    nodes[fn] = {}
    nodes[fn]['x'] = [node[0] for node in nodes_]
    nodes[fn]['y'] = [node[1] for node in nodes_]
data['B']
```

```python
# Looking for best and worst iterations
best = {}
worst = {}

for fn in ['A', 'B']:
    results = [data[fn][i][0] for i in range(2, 300, 3)]
    best[fn] = results.index(min(results))
    worst[fn] = results.index(max(results))
```

```python
# drawing plots
letter = 'B'

nds_l = data[letter][worst[letter]*3]
nds_r = data[letter][worst[letter]*3+1]

#plt.figure()
plt.scatter(nodes[letter]['x'], nodes[letter]['y'], color='#836b6b')

for nds, c in zip([nds_l, nds_r], ['#df2935', '#3772ff']):
    for n in range(len(nds)-2):
        plt.plot([nodes[letter]['x'][int(nds[n])], nodes[letter]['x'][int(nds[n+1])]], 
                 [nodes[letter]['y'][int(nds[n])], nodes[letter]['y'][int(nds[n+1])]],
                  c=c,
                  linewidth=0.7)
    plt.plot([nodes[letter]['x'][int(nds[0])], nodes[letter]['x'][int(nds[len(nds)-2])]], 
             [nodes[letter]['y'][int(nds[0])], nodes[letter]['y'][int(nds[len(nds)-2])]],
              c=c,
              linewidth=0.7)
    plt.scatter([nodes[letter]['x'][int(nds[0])], nodes[letter]['x'][int(nds[len(nds)-2])]], 
                [nodes[letter]['y'][int(nds[0])], nodes[letter]['y'][int(nds[len(nds)-2])]],
                 c=c)
plt.axis(False)
#plt.savefig('imgs/best_kroB100_HNN.png')
```
